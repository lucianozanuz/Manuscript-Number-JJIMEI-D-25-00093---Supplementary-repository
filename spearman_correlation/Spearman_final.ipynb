```python
import pandas as pd

# 1. Carregar o arquivo original
df_raw = pd.read_excel('spearman_corr_nlp_metrics.xlsx', header=[0, 1])

# 2. Definir as métricas e modelos de interesse
models_of_interest = ['gpt-4o-v1', 'gpt-4o-mini-v1',
                      'gpt-4o-v2', 'gpt-4o-mini-v2']

metrics_of_interest = [
    'rouge1', 'rouge2', 'rougeL', 'rougeLsum', 'bleu', 'meteor',
    'bertscore_precision', 'bertscore_recall', 'bertscore_f1', 'bleurt',
    'ada1', 'ada2', 'hf1', 'hf2', 'hf3', 'hf4', 'hf5', 'hf6'
]

# 3. Lista para armazenar as novas linhas
processed_data = []

# Iterar sobre cada processo (linha do dataset original)
for index, row in df_raw.iterrows():
    case_number = row[('case_number', 'Unnamed: 0_level_1')]

    # Para cada processo, extrair os dados de cada modelo
    for model in models_of_interest:
        new_row = {
            'case_number': case_number,
            'model': model
        }

        # Adicionar o valor de cada métrica para aquele modelo
        for metric in metrics_of_interest:
            try:
                # Acessa a tupla (Métrica, Modelo) no MultiIndex
                val = row[(metric, model)]
                new_row[metric] = val
            except KeyError:
                new_row[metric] = None

        processed_data.append(new_row)

# 4. Criar o novo DataFrame
df_final = pd.DataFrame(processed_data)

print("Novo dataset gerado com sucesso!")
print(df_final.head())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# 1. Load the dataset with all merged metrics
df = df_merged_metrics

# 2. Organization of metrics by "Family" for clear analysis
# Including the classification of ada1 and ada2 as semantic embedding metrics
lexical_metrics = ['rouge1', 'rouge2', 'rougeL', 'rougeLsum', 'bleu', 'meteor']
semantic_metrics_bert = ['bertscore_precision', 'bertscore_recall', 'bertscore_f1', 'bleurt']
embedding_distance_openai = ['ada1', 'ada2'] # Embedding distance metrics; lower values indicate better quality
embedding_distance_hf = ['hf1', 'hf2', 'hf3', 'hf4', 'hf5', 'hf6'] # Embedding distance metrics; lower values indicate better quality
human_and_llm_judge_metrics = ['mqm_llm_as_a_judge', 'mqm_human']

all_metrics = (lexical_metrics + semantic_metrics_bert +
               embedding_distance_openai + embedding_distance_hf +
               human_and_llm_judge_metrics)

# Convert to numeric and clean NaN values
df_metrics = df[all_metrics].apply(pd.to_numeric, errors='coerce').dropna()

# 3. Calculate Spearman and Significance
def get_stats(df):
    corr = df.corr(method='spearman')
    pvals = pd.DataFrame(np.zeros((len(df.columns), len(df.columns))),
                         columns=df.columns, index=df.columns)
    for i in df.columns:
        for j in df.columns:
            # spearmanr returns (correlation, p-value)
            _, p = spearmanr(df[i], df[j])
            pvals.loc[i, j] = p
    return corr, pvals

corr_df, p_df = get_stats(df_metrics)

# 4. Professional Heatmap
plt.figure(figsize=(16, 12))
mask = np.triu(np.ones_like(corr_df, dtype=bool))

sns.heatmap(corr_df, annot=True, fmt=".2f", cmap='RdYlBu', mask=mask,
            linewidths=.5, cbar_kws={"label": "Spearman Correlation (ρ)"})

plt.title('Convergence Analysis: NLP Metrics vs. LLM-as-a-Judge vs. Human MQM', fontsize=16)
plt.tight_layout()
plt.savefig('fig_correlation_matrix.png', dpi=300) # Updated filename
plt.show()

# 5. Evidence Table for the Journal
print("\n--- Evidência de Convergência (Métricas NLP, LLM-as-a-Judge e Humana) ---")

# Define embedding distance metrics for interpretation
distance_metrics = embedding_distance_openai + embedding_distance_hf

print(f"\n--- Correlações com 'mqm_llm_as_a_judge' ---")
target_llm = 'mqm_llm_as_a_judge'
for m in ['rougeL', 'bertscore_f1', 'hf1', 'ada1']:
    if m in corr_df.columns and target_llm in corr_df.index:
        rho = corr_df.loc[target_llm, m]
        p = p_df.loc[target_llm, m]
        note = " (Negative rho means better alignment for distance metrics)" if m in distance_metrics else ""
        print(f"{target_llm} vs {m}: ρ={rho:.3f}, p={p:.4e}{note}")

print(f"\n--- Correlações com 'mqm_human' ---")
target_human = 'mqm_human'
for m in ['rougeL', 'bertscore_f1', 'hf1', 'ada1', 'mqm_llm_as_a_judge']:
    if m in corr_df.columns and target_human in corr_df.index:
        rho = corr_df.loc[target_human, m]
        p = p_df.loc[target_human, m]
        note = " (Negative rho means better alignment for distance metrics)" if m in distance_metrics else ""
        print(f"{target_human} vs {m}: ρ={rho:.3f}, p={p:.4e}{note}")
```
